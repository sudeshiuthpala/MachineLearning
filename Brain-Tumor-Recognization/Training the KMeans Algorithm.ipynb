```python
import cv2
from matplotlib import pyplot as plt
img=cv2.imread('test_images/0 (239).jpg',0)#load images from test image that allimages
print(img.shape)
plt.imshow(img,cmap='gray')
#pixcels 512*512 eka eka clusters walata kadanawa
#e clusters walin brain tumer ekata adala kotasa mokkda balanawa
#feature pixcel eke color eka =>color segmentation(gray scale) features 1 0-255
#RGB nm 3 k thiyanwa(0-255,0-255,0-255)
#reshape
height,width=img.shape
imgFlatten=img.reshape(height*width,1)#column vector ekata flat karanawa
print(imgFlatten.shape)
print(imgFlatten[50000:51000])
#one row represent the one pixcel and value is color of pixcel. 
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=5)
model.fit(imgFlatten)
labels=model.labels_
#1/0
print(len(labels))#1D
labels2D=labels.reshape(height,width)#2D
print(labels2D.shape)
print(labels[50000:51000])
```

```python
from matplotlib import pyplot as plt
# cluster ekata adala hoyaganna color eka
#tumer detect cluster number=4
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
import joblib

joblib.dump(model,'brain_tumor_kmeans.sav')
```
