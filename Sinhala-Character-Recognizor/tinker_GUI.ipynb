```python

import tkinter as tk
#pip install pillow
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import cv2
import os
import joblib

model=joblib.load('sinhala-character-knn.sav')

width=500
height=500

count=0

label_dict={0:'අ',1:'එ',2:'ඉ',3:'උ'}

font_btn='Helvetica 20 bold'
font_label='Helvetica 22 bold'
#main window 1st step
win=tk.Tk()

#6th step
#tkinter event B1-Mortion
def event_function(event):
    x=event.x
    y=event.y
    
    x1=x-30
    x2=x+30
    
    y1=y-30
    y2=y+30
    
    canvas.create_oval((x1,y1,x2,y2),fill='black')#create a circle# design the pattern #After that we can read
    
    #7th step
    img_draw.ellipse((x1,y1,x2,y2),fill='black')
    #print(x,y)
    #print('Here')
    

#8th step
def save():
    #image eka convert to numpy array
    #path must be name change
    
    global count
    
    img_array=np.array(img)
    path=os.path.join('data',str(count)+'.jpg')
    #path=data/0.jpg
    cv2.imwrite(path,img_array)
    count=count+1

#9th step    
def clear():
    #replace karanawa parana eka ghilla
    global img,img_draw

    canvas.delete('all')
    #local
    img=Image.new('RGB',(width,height),(255,255,255))
    img_draw=ImageDraw.Draw(img)  
    
    label_status.config(text='PREDICTED CHARACTER: NONE')
    
 #10th step   
def predict():
    #convert to numpy array,convert to grey,resize,flatting(1d karanawa),
    img_array=np.array(img)
    
    img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2GRAY) #converting into a gray image
    img_array=cv2.resize(img_array,(8,8)) #resizing into 8x8  
    img_array=np.reshape(img_array,(1,64))  #reshaping into 1x64 
    
    #predict karanawa
    #result=model.predict(img_array)
    #beacause multi dimentional so access 0th one 
    result=model.predict(img_array)[0]
    #print(result)
    print(label_dict[result])
    label=label_dict[result]
    
    label_status.config(text='PREDICTED CHARACTER:'+label)
    
# add canvan widget-2nd step
canvas=tk.Canvas(win,width=width,height=height,bg='white')#(mainwindow name,width,height,color)
#3-methods
#grid= can defind by rows and columns
canvas.grid(row=0,column=0,columnspan=4)#canvas eka place karanawa

#add button-3rd step

button_save=tk.Button(win,text='SAVE',bg='green',fg='white',font=font_btn,command=save)
button_save.grid(row=1,column=0)

button_predict=tk.Button(win,text='PREDICT',bg='blue',fg='white',font=font_btn,command=predict)
button_predict.grid(row=1,column=1)

button_clear=tk.Button(win,text='CLEAR',bg='yellow',fg='white',font=font_btn,command=clear)
button_clear.grid(row=1,column=2)

button_exit=tk.Button(win,text='EXIT',bg='red',fg='white',font=font_btn,command=win.destroy)
button_exit.grid(row=1,column=3)

#add label-4th step
label_status=tk.Label(win,text='PREDICTED CHARACTER: NONE',bg='white',font=font_label)
label_status.grid(row=2,column=0,columnspan=4)

#5th Step
#draw in canvas
#Button1 move weddi event _function eka wenna ona
#canvas.bind('<Button-1>',event_function)#left one inmouse
#canvas.bind('<Button-2>',event_function)#scroll one inmouse
canvas.bind('<B1-Motion>',event_function)

#7th step
#create a same to same image that user draw in canvas 
img=Image.new('RGB',(width,height),(255,255,255))
img_draw=ImageDraw.Draw(img)

#1ststep
win.mainloop()#display the main window
```
